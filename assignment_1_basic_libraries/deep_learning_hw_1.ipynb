```python
# Audio specgram task (task 7)

# I used a website to generate a 1k and 10k Hz sine wave:
# https://www.audiocheck.net/audiofrequencysignalgenerator_sinetone.php
# I also used another plotting measure to make sure the results were correct
# I didn't know how to add the sine_10k.wav file to a gist repo, I simply had it in the same
# folder as the jupyter notebook.

#used imports again so this can be ran separately
import matplotlib.pyplot as plt
from scipy.fftpack import fft
from scipy.io import wavfile # get the api

#fs, data = wavfile.read('sine_1k.wav') # load the data
fs, data = wavfile.read('sine_10k.wav') # load the data

dt = 1/fs
t = np.arange(0, 10, dt)
data = np.asarray(data)

print(fs)
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(7, 7))


axes[0].set_title("Signal")
#1 kHz -> N = 1000, 10kHz-> 300
N=300
#plot sine in time
axes[0].plot(t[:N], data[:N], color='C0')
axes[0].set_xlabel("Time")
axes[0].set_ylabel("Amplitude")
#plot magnitude spectrum
axes[1].set_title("Magnitude Spec")
axes[1].magnitude_spectrum(data[:500], Fs=fs, color='C1')
```

```python
#Letter counter task (task 8)
import requests
from bs4 import BeautifulSoup as bs
from collections import Counter
import matplotlib.pyplot as plt
import pandas as pd
#plt.rcParams.update({'figure.autolayout': True})
url = "https://www.snopes.com/fact-check/facebook-ai-developed-own-language/"

#download html document
response = requests.get(url)
print("status code: " + str(response.status_code))
content = response.content
soup = bs(content, "lxml")
```

```python
%%capture
#remove unnecessary tags:
# To mostly just keep the article texts i removed head and script tags from the document, the reason i remove script tags
# after head is because not all script tages were inside the head tags.

[s.extract() for s in soup('head')]
[s.extract() for s in soup('script')]

#pretty_html = soup.prettify()
#print(pretty_html)
```

```python
# after most of the clutter (scripts, styling strings) removed I print the text of the html document which will mostly be
# be the actual article with a lot of white spaces.
print(soup.text) # this prints the document without any html tags
```

```python
# Here I move on to the counting of letters in the string that doesn't contain any of the html tags
# use lower case for all letters, so they aren't counted separately
str_lower = soup.text.lower()
#print(string)
cntr = Counter(str_lower)
#remove non letters from the original dictionary returned by the Counter()
cntr={key: value for key, value in cntr.items() if key.isalpha()}
#get datafram and order the values 
df = pd.DataFrame.from_dict(cntr, orient='index')
df = df.sort_values(by=[0], ascending=False)
print(cntr)
# here I use kind = 'bar' not 'hist' because the array already contains the frequency of the occurances 
df.plot(kind='bar',title="Letter Histogram", figsize=(15,6), rot = 0, fontsize = 14)
plt.show()
```
